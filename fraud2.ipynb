```python
import numpy as np
import pandas as pd
df = pd.read_csv(r'Job_Frauds.csv', encoding="ISO-8859-1")
```

```python
df=df[['Job_Description','Requirements','Fraudulent']]
df.head()
```

```python
###Drop Nan Values
df=df.dropna()
```

```python
df.shape
```

```python
## Get the Independent Features

x=df.drop('Fraudulent',axis=1)
```

```python
## Get the Dependent features
y=df['Fraudulent']
```

```python
y.value_counts()
```

```python
import tensorflow as tf
tf .__version__
```

```python
# importing tensorflow dependancies
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Bidirectional
from tensorflow.keras.layers import Dropout
```

```python
voc_size = 5000
```

```python
# One-Hot Presentation
message = x.copy()
```

```python
message['Job_Description'][1]
```

```python
message.reset_index(inplace=True)
```

```python
#eliminating stopwords 

import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
# using PortStemmer

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(message)):
    review = re.sub('[^a-zA-Z]', ' ', message['Job_Description'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
onehot_repr=[one_hot(words,voc_size)for words in corpus] 
onehot_repr[1]
```

```python
sent_length=40
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
embedded_docs[0]
```

```python

## Creating model
embedding_vector_features=50
model1=Sequential()
model1.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model1.add(Bidirectional(LSTM(100))) ##Just add bidirectional!!, except it would just behave as normal LSTM Model
model1.add(Dropout(0.3))
model1.add(Dense(1,activation='sigmoid'))
model1.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model1.summary())
```

```python
len(embedded_docs),y.shape
```

```python
import numpy as np
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
X_final[1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.25, random_state=32)
```

```python
model1.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=12,batch_size=64)
```

```python
# Model Performance & Accuracy 
y_pred = np.argmax(model1.predict(X_test), axis=-1)
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
